```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axes = plt.subplots(1, 2, figsize=(10, 3))

ax = axes[0]
ax.plot(x, y1)
ax.set_title('Sin')
ax.set_xlabel('x')
ax.set_ylabel('y')


ax = axes[1]
ax.plot(x, y2)
ax.set_title('Cos')
ax.set_xlabel('x')
```
