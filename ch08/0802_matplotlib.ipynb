```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 7, 2, 9, 5])  

plt.scatter(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, y)
```

```python
plt.scatter(x, y, label='散点')
plt.plot(x, y, label='连线')
plt.title('散点和连线')
plt.legend()
plt.xlabel('x轴')
plt.ylabel('y轴')
```

```python
种类 = ['A', 'B', 'C', 'D']
库存 = np.array([32, 40, 32, 28])

fig, ax = plt.subplots(figsize=(8, 4), dpi=100)
ax.bar(种类, 库存, width=0.8, label='月初库存')

ax.set_xlabel('类别')
ax.set_ylabel('库存')
```

```python
种类 = ['A', 'B', 'C', 'D']
月初库存 = [20, 35, 30, 25]
月末库存 = [32, 40, 32, 28]

fig, ax = plt.subplots(figsize=(8, 4), dpi=100)
cat_axis = np.arange(len(种类))
ax.bar(cat_axis - 0.2, 月初库存, width=0.4, label='月初库存')
ax.bar(cat_axis + 0.2, 月末库存, width=0.4, label='月末库存')

ax.set_xticks(cat_axis, 种类)
ax.set_xlabel('类别')
ax.set_ylabel('库存')
ax.legend(loc='upper right');
```

```python
fig, ax = plt.subplots(figsize=(8, 4), dpi=100)
cat_axis = np.arange(len(种类))
ax.barh(cat_axis - 0.2, 月初库存, height=0.4, label='月初库存')
ax.barh(cat_axis + 0.2, 月末库存, height=0.4, label='月末库存')

ax.set_yticks(cat_axis, 种类)
ax.set_ylabel('类别')
ax.set_xlabel('库存')
ax.legend(loc='upper right');
```

```python
种类 = ['A', 'B', 'C', 'D']
库存 = np.array([32, 40, 32, 28])

fig, ax = plt.subplots(figsize=(5, 3))
ax.pie(库存, labels=种类);
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.pie(库存, wedgeprops=dict(width=0.5), labels=种类);
```

## 绘制函数曲线


```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y)
ax.set_xlabel('x'); ax.set_ylabel('y');
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axes = plt.subplots(1, 2, figsize=(10, 3))

ax = axes[0]
ax.plot(x, y1)
ax.set_title('Sin')
ax.set_xlabel('x')
ax.set_ylabel('y')


ax = axes[1]
ax.plot(x, y2)
ax.set_title('Cos')
ax.set_xlabel('x')
```
